```python
import numpy as np
import matplotlib.pyplot as plt

step = 100
people = 1000
a,b,tr1,tr2 = [], [], [0], [0]

#Random Walk 1:
trajectory_mode = True #第一次循環紀錄移動路徑
for i in range(0, people):
    locate = 0 #換人走之後初始位置歸零
    for j in range(0, step):
        x = np.random.random() - 1/2
        locate += x
        if trajectory_mode == True:
            tr1.append(locate)
    trajectory_mode = False #第一個人走完了，終止路線追蹤
    a.append(locate)

#Random Walk 2:
trajectory_mode = True
for j in range(0,people):
    locate = 0
    for i in range(0,step):
        x = np.random.randint(0,2)
        if x == 0:
            locate = locate - 1
        else:
            locate = locate + 1    
        if trajectory_mode == True:
            tr2.append(locate)
    trajectory_mode = False
    b.append(locate)

plt.style.use("dark_background")

plt.title('1D Random Walk Track 1')
plt.plot(tr1)
plt.show
```

```python
plt.title('1D Random Walk Histogram 1')
plt.hist(a,30)
plt.show
```

```python
plt.title('1D Random Walk Track 2')
plt.plot(tr2)
plt.show
```

```python
plt.title('1D Random Walk Histogram 2')
plt.hist(b,30)
plt.show
```

```python
#二維 Random Walk, dx, dy = (-1/2, 1/2)
step = 500
people = 600
#軌跡紀錄
pos1 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    dx = np.random.random() - 1/2
    dy = np.random.random() - 1/2
    pos1[i,0] = pos1[i-1,0] + dx
    pos1[i,1] = pos1[i-1,1] + dy

#終點紀錄
endpoint = np.zeros([people,3]) #第三列用來紀錄終點起點距離
for j in range(0,people):
    locate = [0,0]
    for i in range(0,step):
        dx = np.random.random() - 1/2
        dy = np.random.random() - 1/2
        locate[0] += dx
        locate[1] += dy
    endpoint[j,0:2] = locate
    endpoint[j,2] = endpoint[j,0] ** 2 + endpoint[j,1] ** 2 #終點距離

plt.style.use('dark_background')
plt.title('2D Random dxdy Track Mode')
plt.plot(pos1[:,0], pos1[:,1])
plt.show
```

```python
plt.title('2D Random dxdy Endpoint Mode')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```

```python
#二維 Random Walk dxdy +-1 mode
step = 500
prople = 600
pos2 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    x = np.random.randint(0,2) #產生1/2機率
    y = np.random.randint(0,2)
    if x == 0:
        pos2[i,0] = pos2[i-1,0] + 1
    else:
        pos2[i,0] = pos2[i-1,0] - 1
    if y == 0:
        pos2[i,1] = pos2[i-1,1] + 1
    else:
        pos2[i,1] = pos2[i-1,1] - 1

endpoint = np.zeros([people,3])
for j in range(0,people):
    locate = [0,0]
    for i in range(0,step):
        x = np.random.randint(0,2) #產生1/2機率
        y = np.random.randint(0,2)
        if x == 0:
            locate[0] += 1
        else:
            locate[0] -= 1
        if y == 0:
            locate[1] += 1
        else:
            locate[1] -= 1
    endpoint[j,0:2] = locate
    endpoint[j,2] = endpoint[j,0] ** 2 + endpoint[j,1] ** 2 #終點距離
plt.style.use('dark_background')
plt.title('2D Random dxdy +-1 Mode')
plt.plot(pos2[:,0], pos2[:,1])
plt.show
```

```python
plt.title('2D Random dxdy +-1 Endpoint Mode')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy  +-1 Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```

```python
#二維 Random Walk 固定步長
step = 500
people = 600
pos3 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    theta = np.random.random() * 2 * np.pi
    pos3[i,0] = pos3[i-1,0] + np.cos(theta)
    pos3[i,1] = pos3[i-1,1] + np.sin(theta)

endpoint = np.zeros([people,3])
for j in range(0,people):
    locate = [0,0]
    for i in range(1,step):
        theta = np.random.random() * 2 * np.pi
        locate[0] += np.cos(theta)
        locate[1] += np.sin(theta)
    endpoint[j,0:2] = locate
    endpoint[j,2] = endpoint[j,0] ** 2 + endpoint[j,1] ** 2 #終點距離
plt.style.use('dark_background')
plt.title('2D Ramdom Walk theta Track Mode')
plt.plot(pos3[:,0], pos3[:,1])
plt.show
```

```python
plt.title('2D Random dxdy theta Endpoint Mode')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy theta Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```
