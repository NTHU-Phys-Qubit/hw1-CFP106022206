```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
step = 200
people = 1000
a,b,tr1,tr2 = [], [], [0], [0]

#Random Walk 1:
trajectory_mode = True #第一次循環紀錄移動路徑
for i in range(0, people):
    locate = 0 #換人走之後初始位置歸零
    for j in range(0, step):
        x = np.random.random() - 1/2
        locate += x
        if trajectory_mode == True:
            tr1.append(locate)
    trajectory_mode = False #第一個人走完了，終止路線追蹤
    a.append(locate)

#Random Walk 2:
trajectory_mode = True
for j in range(0,people):
    locate = 0
    for i in range(0,step):
        x = np.random.randint(0,2)
        if x == 0:
            locate = locate - 1
        else:
            locate = locate + 1    
        if trajectory_mode == True:
            tr2.append(locate)
    trajectory_mode = False
    b.append(locate)

plt.style.use("dark_background")

plt.title('1D Random Walk Track 1')
plt.plot(tr1)
plt.show
```

```python
plt.title('1D Random Walk Histogram 1')
plt.hist(a,30)
plt.show
```

```python
plt.title('1D Random Walk Track 2')
plt.plot(tr2)
plt.show
```

```python
plt.title('1D Random Walk Histogram 2')
plt.hist(b,30)
plt.show
```
