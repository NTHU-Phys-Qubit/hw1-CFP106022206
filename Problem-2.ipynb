# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
#二維 Random Walk, dx, dy = (-1/2, 1/2)
step = 1000
people = 600
#軌跡紀錄
pos1 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    pos1[i,0] = pos1[i-1,0] + np.random.random() - 1/2
    pos1[i,1] = pos1[i-1,1] + np.random.random() - 1/2

#終點紀錄
endpoint = np.zeros([people,3]) #第三列用來紀錄終點起點距離
for j in range(0,people):
    locate = [0,0]
    for i in range(0,step):
        locate[0] += np.random.random() - 1/2
        locate[1] += np.random.random() - 1/2
    endpoint[j,0:2] = locate
    endpoint[j,2] = np.sqrt(endpoint[j,0] ** 2 + endpoint[j,1] ** 2) #終點距離
plt.style.use('dark_background')
plt.axis('equal')
plt.title('2D Random dxdy Track Mode')
plt.plot(pos1[:,0], pos1[:,1])
plt.show
```

```python
plt.title('2D Random dxdy Endpoint Mode')
plt.axis('equal')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```

```python
#二維 Random Walk dxdy +-1 mode
step = 1000
prople = 600
pos2 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    pos2[i,0] = pos2[i-1,0] + 2*np.random.randint(0,2) - 1
    pos2[i,1] = pos2[i-1,1] + 2*np.random.randint(0,2) - 1

endpoint = np.zeros([people,3])
for j in range(0,people):
    locate = [0,0]
    for i in range(0,step):
        locate[0] += 2*np.random.randint(0,2) -1
        locate[1] += 2*np.random.randint(0,2) -1
    endpoint[j,0:2] = locate
    endpoint[j,2] = np.sqrt(endpoint[j,0] ** 2 + endpoint[j,1] ** 2) #終點距離
plt.style.use('dark_background')
plt.axis('equal')
plt.title('2D Random dxdy +-1 Mode')
plt.plot(pos2[:,0], pos2[:,1])
plt.show
```

```python
plt.title('2D Random dxdy +-1 Endpoint Mode')
plt.axis('equal')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy  +-1 Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```

```python
#二維 Random Walk 固定步長
step = 1500
people = 600
pos3 = np.zeros([step,2]) #位置紀錄矩陣

for i in range(1,step):
    theta = np.random.random() * 2 * np.pi
    pos3[i,0] = pos3[i-1,0] + np.cos(theta)
    pos3[i,1] = pos3[i-1,1] + np.sin(theta)

endpoint = np.zeros([people,3])
for j in range(0,people):
    locate = [0,0]
    for i in range(1,step):
        theta = np.random.random() * 2 * np.pi
        locate[0] += np.cos(theta)
        locate[1] += np.sin(theta)
    endpoint[j,0:2] = locate
    endpoint[j,2] = np.sqrt(endpoint[j,0] ** 2 + endpoint[j,1] ** 2) #終點距離
plt.style.use('dark_background')
plt.axis('equal')
plt.title('2D Ramdom Walk theta Track Mode')
plt.plot(pos3[:,0], pos3[:,1])
plt.show
```

```python
plt.title('2D Random dxdy theta Endpoint Mode')
plt.axis('equal')
plt.plot(endpoint[:,0],endpoint[:,1],'.')
plt.show
```

```python
plt.title('2D Random dxdy theta Endpoint Histogram')
plt.hist(endpoint[:,2],50)
plt.show
```
